```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip list
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('titanic')
data.head()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x = 'day', y= 'tip', data = tips)
```

```python
sns.barplot(x = 'day', y= 'tip', data = tips)
plt.xlabel('Days')
plt.ylabel('Tip')
plt.xticks(rotation = 45)
```

```python
sns.barplot(x = 'day', y= 'tip', data = tips, hue = 'sex')
```

```python
sns.barplot(x = 'day', y= 'tip', data = tips, hue = 'sex', palette = 'spring')
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
```

```python
tips.describe()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', estimator = np.median, ci = 0)
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'])
# Curve is Kernel Density Estimation
```

```python
plt.hist(tips['total_bill'])
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], hist = False)
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], kde = False)
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], color = 'red')
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], color = 'red', vertical = True)
```
