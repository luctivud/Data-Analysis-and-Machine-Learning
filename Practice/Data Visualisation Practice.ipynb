## Bar Graph

```python
lang = np.array(['C', 'C++', 'Python', 'R', 'Java'])
students = np.array([12, 47, 51, 2, 31])
plt.bar(lang, students)
plt.show()
```

```python
plt.bar(lang, students, color = 'red', width = 0.5)
# width mein 1 pr overlapping hogi
plt.show()
```

```python
N = 5
men = np.array([10, 22, 32, 45, 22])
women = np.array([4, 21, 44, 21, 32])
var = np.arange(1, N+1)
width = 0.4
plt.bar(var, men, width, label = 'Men', color = 'black')
plt.bar(var+width, women, width, label = 'Women', color = 'red')
plt.grid(True)
plt.legend()
plt.ylabel('Participants')
plt.xlabel("Day")
plt.title('Number of participants in male and female')
```

```python
N = 5
men = np.array([10, 22, 32, 45, 22])
women = np.array([4, 21, 44, 21, 32])
var = np.arange(1, N+1)
width = 0.4
plt.bar(var-(width/2), men, width, label = 'Men', color = 'black')
plt.bar(var+(width/2), women, width, label = 'Women', color = 'red')
plt.grid(True)
plt.legend()
plt.ylabel('Participants')
plt.xlabel("Day")
plt.title('Number of participants in male and female')
```
