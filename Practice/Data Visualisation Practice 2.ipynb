### Scatter Plot

```python
x = np.linspace(0, 10, 30)
plt.plot(x, np.sin(x), 'D', color = 'k', markersize = 5)
```

```python
N = 60
g1 = (np.random.rand(N), np.random.rand(N))
g2 = (np.random.rand(N), 0.5 * np.random.rand(N))
g3 = (np.random.rand(N), 0.3 * np.random.rand(N))
data = (g1, g2, g3)
labels = ('coffee', 'tea', 'water')
color = ('red', 'green', 'blue')
for data, color, labels in zip(data, color, labels):
    x, y = data
    plt.scatter(x, y)
```

```python
N = 10
g1 = (np.random.rand(N), np.random.rand(N))
g2 = (np.random.rand(N), 0.5 * np.random.rand(N))
data = (g1, g2)
labels = ('coffee', 'tea')
color = ('red', 'green')
for data, color, labels in zip(data, color, labels):
    x, y = data
    plt.scatter(x, y)
```

```python
lang = ['Julia', 'Perl', 'GoLang', 'Scala']
sizes = ['210', '123', '321', '333']
color = ['gold', 'cyan', 'fuchsia', 'coral']
explode = [0.1, 0, 0, 0]
plt.pie(sizes, labels = lang, colors = color, explode = explode, autopct = '%1.1f%%', pctdistance = 0.5, shadow = True)
plt.show()
```

```python
age = np.random.randint(0, 100, size = 70)
bins = np.arange(0, 100, 10)
plt.xlabel('Age')
plt.ylabel('Person')
plt.title('No. of person in age group')
plt.hist(age, bins, color = 'm')
```

```python
bloodmen = np.random.randint(90, 150, size = 10)
bloodwomen = np.random.randint(90, 150, size = 10)
plt.hist([bloodmen, bloodwomen], bins = np.arange(90, 150, 10), color = ['red', 'blue'], label = ['Men', 'Women'])
plt.legend()
```

```python
plt.hist([bloodmen, bloodwomen], bins = 10, color = ['red', 'blue'], label = ['Men', 'Women'])
plt.legend()
```

```python
plt.hist([bloodmen, bloodwomen], bins = np.arange(90, 150, 10), color = ['red', 'blue'], label = ['Men', 'Women'], orientation = 'horizontal')
plt.legend()
```
